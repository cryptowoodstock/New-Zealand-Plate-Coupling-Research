```python
from shapely.geometry import Point
from shapely.geometry import Polygon
import numpy as np

with open('NZ_all_m0mw.txt') as f:
# with open('meca70mag.txt') as f:
    array = [[x for x in line.split()] for line in f]
cmt_data = np.array(array)   

mrrsum = []
mttsum = []
mppsum = []
mrtsum = []
mrpsum = []
mtpsum = []
depavg = [] 

zone_cmt = []


#zone 1
zone = Polygon([(160,-47),(166,-47),(166,-50),(160,-50)])

for line in cmt_data:
    coordinates = Point(float(line[0]),float(line[1]))
  
    if coordinates.within(zone):
        zone_cmt.append(line[0:13])

mrr = 0
mtt = 0
mpp = 0
mrt = 0
mrp = 0 
mtp = 0
dep = 0


for k in range(len(zone_cmt)):
    dep += float(zone_cmt[k][2])
    mrr += float(zone_cmt[k][3])*(10**(float(zone_cmt[k][9])-7))
    mtt += float(zone_cmt[k][4])*(10**(float(zone_cmt[k][9])-7))
    mpp += float(zone_cmt[k][5])*(10**(float(zone_cmt[k][9])-7))
    mrt += float(zone_cmt[k][6])*(10**(float(zone_cmt[k][9])-7))
    mrp += float(zone_cmt[k][7])*(10**(float(zone_cmt[k][9])-7))
    mtp += float(zone_cmt[k][8])*(10**(float(zone_cmt[k][9])-7))
  
    
mrrsum.append(mrr)
mttsum.append(mtt)
mppsum.append(mpp)
mrtsum.append(mrt)
mrpsum.append(mrp)
mtpsum.append(mtp)
depavg.append(dep/len(zone_cmt))


```

```python
longitude = [(164.5),(166.5),(169.5),(172.5),(172.5),(176),(176),(178.5),(179),(-179.5),(-178.5),(-177.5),(179.5)]
latitude = [(-48.5),(-45.5),(-43.5),(-43.5),(-42),(-41),(-39),(-38.5),(-35),(-31),(-33),(-34.5),(-40)]
assert len(longitude) == len(latitude)
smallzonedata = np.zeros((13,12), dtype=object)
count = 0

for i in range(len(longitude)):
        smallzonedata[i,0] = longitude[i]
        smallzonedata[i,1] = latitude[i]
        smallzonedata[i,2] = depavg[i]
        smallzonedata[i,3] = mrrsum[i]/(10**20)
        smallzonedata[i,4] = mttsum[i]/(10**20)
        smallzonedata[i,5] = mppsum[i]/(10**20)
        smallzonedata[i,6] = mrtsum[i]/(10**20)
        smallzonedata[i,7] = mrpsum[i]/(10**20)
        smallzonedata[i,8] = mtpsum[i]/(10**20)
        smallzonedata[i,9] = 27
        smallzonedata[i,10] = "X"
        smallzonedata[i,11] = "Y"
        count += 1
        
f = open('smallzonegrpall.txt','w')  
for row in smallzonedata:
    for column in row:
        f.write("%s"" " % column)
    f.write("\n") 
f.close()
```
