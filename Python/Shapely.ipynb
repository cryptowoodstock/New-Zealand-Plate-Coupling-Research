```python
#zone 11
from shapely.geometry import Point
from shapely.geometry import Polygon
import numpy as np

zone11 = Polygon([(166.9,-46.5),(169,-45.3),(167,-43.8),(163.6,-45.6)])

zone11_cmt = []

with open('meca.txt') as f:
    array = [[x for x in line.split()] for line in f]
cmt_data = np.array(array)   

for line in cmt_data:
    coordinates = Point(float(line[0]),float(line[1]))
    
    if coordinates.within(zone11):
        zone11_cmt.append(line[0:13])
        
f = open('zone11.txt','w')  
for row in zone11_cmt:
    for column in row:
        f.write("%s"" " % column)
    f.write("\n") 
f.close()

mrr = 0
mtt = 0
mpp = 0
mrt = 0
mrp = 0 
mtp = 0

for i in range(len(zone11_cmt)):
    mrr += float(zone11_cmt[i][3])*(10**(float(zone11_cmt[i][9])-7))
    mtt += float(zone11_cmt[i][4])*(10**(float(zone11_cmt[i][9])-7))
    mpp += float(zone11_cmt[i][5])*(10**(float(zone11_cmt[i][9])-7))
    mrt += float(zone11_cmt[i][6])*(10**(float(zone11_cmt[i][9])-7))
    mrp += float(zone11_cmt[i][7])*(10**(float(zone11_cmt[i][9])-7))
    mtp += float(zone11_cmt[i][8])*(10**(float(zone11_cmt[i][9])-7))

print('mrr = ', mrr)
print('mtt = ', mtt)   
print('mpp = ', mpp)
print('mrt = ', mrt)
print('mrp = ', mrp)
print('mtp = ', mtp)
```
