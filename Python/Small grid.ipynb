```python
from shapely.geometry import Point
from shapely.geometry import Polygon
import numpy as np

longitude = 160
orig_long = longitude
latitude = -30
long = []
lat = []
coor = []

for j in range(21):
    coor.append([])
    for i in range(21):
        long.append(longitude)
        lat.append(latitude)
        coor[j].append([longitude,latitude])
        longitude += 1
    longitude = orig_long
    latitude -= 1 

#print(Polygon([(coor[0][0]),(coor[0][1]),(coor[1][1]),(coor[1][0])]))
#zone = Polygon([(long[i],latitude),(longitude + 1,latitude),(longitude,latitude - 1),(longitude + 1,latitude - 1)])
#zone = Polygon([(coor[i][j]),(coor[i+1][j]),(coor[i][j-1]),(coor[i+1][j-1])])


```

```python
with open('meca70.txt') as f:
    array = [[x for x in line.split()] for line in f]
cmt_data = np.array(array)   

mrrsum = []
mttsum = []
mppsum = []
mrtsum = []
mrpsum = []
mtpsum = []

for i in range(20):
    for j in range(20):
        zone_cmt = []
        for line in cmt_data:
            coordinates = Point(float(line[0]),float(line[1]))
            zone = Polygon([(coor[i][j]),(coor[i][j+1]),(coor[i+1][j+1]),(coor[i+1][j])])
            if coordinates.within(zone):
                zone_cmt.append(line[0:13])

        mrr = 0
        mtt = 0
        mpp = 0
        mrt = 0
        mrp = 0 
        mtp = 0

        for k in range(len(zone_cmt)):
            mrr += float(zone_cmt[k][3])*(10**(float(zone_cmt[k][9])-7))
            mtt += float(zone_cmt[k][4])*(10**(float(zone_cmt[k][9])-7))
            mpp += float(zone_cmt[k][5])*(10**(float(zone_cmt[k][9])-7))
            mrt += float(zone_cmt[k][6])*(10**(float(zone_cmt[k][9])-7))
            mrp += float(zone_cmt[k][7])*(10**(float(zone_cmt[k][9])-7))
            mtp += float(zone_cmt[k][8])*(10**(float(zone_cmt[k][9])-7))
        mrrsum.append(mrr)
        mttsum.append(mtt)
        mppsum.append(mpp)
        mrtsum.append(mrt)
        mrpsum.append(mrp)
        mtpsum.append(mtp)
```

```python
smallzonedata = []
count = 0

for i in range(400):
    if (mrrsum[i] != 0 and mttsum[i] != 0 and mppsum[i] != 0):
        mod = i % 20
        div = int(i/20)
        smallzonedata.append([])
        smallzonedata[count].append(coor[div][mod][0]+0.5)
        smallzonedata[count].append(coor[div][mod][1]-0.5)
        smallzonedata[count].append(35)
        smallzonedata[count].append(mrrsum[i]/(10**20))
        smallzonedata[count].append(mttsum[i]/(10**20))
        smallzonedata[count].append(mppsum[i]/(10**20))
        smallzonedata[count].append(mrtsum[i]/(10**20))
        smallzonedata[count].append(mrpsum[i]/(10**20))
        smallzonedata[count].append(mtpsum[i]/(10**20))
        smallzonedata[count].append(27)
        smallzonedata[count].append('X')
        smallzonedata[count].append('Y')
        count += 1
        
f = open('smallzone.txt','w')  
for row in smallzonedata:
    for column in row:
        f.write("%s"" " % column)
    f.write("\n") 
f.close()
```
